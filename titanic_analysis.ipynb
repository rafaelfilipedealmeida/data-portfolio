# Titanic Dataset Analysis
## Questions I'm answering:
1. What was the survival rate by gender?
2. What was the survival rate by passenger class?
3. What was the average age of survivors vs non-survivors?

1. What was the survival rate by gender?


```python
df.groupby("Sex")["Survived"].mean()
```

**Observation:** 74% of women survived compared to only 19% of men.
Gender was the strongest predictor of survival.

2. What was the survival rate by passenger class?

```python
df.groupby("Pclass")["Survived"].mean()
```

**Observation:** First class passengers had a 63% survival rate compared to 24% in third class.
Socioeconomic status also played a major factor in survival.

3. What was the average age of survivors vs non-survivors?

```python
df.groupby("Survived")["Age"].mean()
```

**Observation:** Survivors had an average age of 28 years vs 30 years for non-survivors.
Age had minimal impact compared to gender and class.

##Visualizations

```python
import matplotlib.pyplot as plt

df.groupby("Sex")["Survived"].mean().plot(kind="bar",color=["salmon", "steelblue"])
plt.title("Survival Rate by Gender")
plt.ylabel("Survival Rate")
plt.xlabel("Gender")
plt.show()
```

```python
import matplotlib.pyplot as plt

df.groupby("Pclass")["Survived"].mean().plot(kind="bar",color=["salmon", "steelblue","darkgreen"])
plt.title("Survival Rate by Passenger Class")
plt.ylabel("Survival Rate")
plt.xlabel("Passenger Class")
plt.show()
```

```python
import matplotlib.pyplot as plt

df.groupby("Survived")["Age"].mean().plot(kind="bar",color=["salmon","steelblue"])
plt.title("Average Age by Survival")
plt.ylabel("Average Age")
plt.xlabel("Survived")
plt.show()
```
